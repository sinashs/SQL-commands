```python
# Show number of loans per month 
# total amount of all loans per month
#average amount of each loan per month
#smallest loan size per month
#largest loan size per month
#order by most recent month first
```

```python
%%sql
SELECT DATE_PART('month', created) AS month, COUNT(id) AS Num_loans, SUM(amount) AS Total_amount, AVG(amount) AS AVG_amount, MIN(amount), MAX(amount)
FROM loans
GROUP BY DATE_PART('month', created)
ORDER BY 1
```

We want to know the loan `amount`,  `user_id`, `user.created` and the `transaction_fee` charged for the **transaction of type `disbursement`** .

Order these results by the date of the customer account created - going by newest first

```python
%%sql
SELECT l.amount AS Loan_amount, l.user_id AS l_UserID, t.user_id AS t_UserID, u.created AS User_created, t.transaction_fee, 
       t.t_type
FROM transactions AS t , loans AS l , users AS u
WHERE u.id = l.user_id AND t.loan_id = l.id 
AND t.t_type = 'disbursement'
ORDER BY u.created DESC
```

```python
%%sql
DROP VIEW IF EXISTS vwfulltransaction
```

```python
%%sql
CREATE VIEW vwfulltransaction AS 
SELECT created , DATE_PART('dow', created ) AS weekday , COUNT(id), SUM(transaction_fee) AS total_transactionfee
    FROM transactions
    GROUP BY 1 
```

```python
%%sql
SELECT * FROM vwfulltransaction
LIMIT 4
```

```python
%%sql
SELECT *
FROM vwfulltransaction t1 , vwfulltransaction t2
WHERE t1.total_transactionfee > t2.total_transactionfee
AND EXTRACT('doy' FROM t1.created) -  EXTRACT('doy'FROM t2.created) = 1
```

```python
# Show total number of transactionfee per month 
```

```python
%%sql 
SELECT DATE_PART('month', created), SUM(count) AS Num_transaction, SUM(total_transactionfee) AS Total_trasanction_fee 
FROM vwfulltransaction
GROUP BY 1
ORDER BY 1
```

```python
# Show the Running_Total_amount, Total amount and percentage total for diffrenet loan status
```

```python
%%sql
SELECT loan_status, 
       amount,
       created,
       ROW_NUMBER() OVER (PARTITION BY loan_status ORDER BY created) AS Row_Num, 
       SUM(amount)  OVER (PARTITION BY loan_status ORDER BY created) AS Running_Total_amount,
       SUM(amount)  OVER (PARTITION BY loan_status ) AS Total_amount,
       amount*100/SUM(amount) OVER (PARTITION BY loan_status) AS Pct_total
       
FROM loans
WHERE id IS NOT NULL
ORDER BY 1
```

```python
%%sql
SELECT l.loan_status AS loan_status, t.t_type AS transaction_type, COUNT(l.id) AS Num_loans, 
       SUM(l.amount) AS Total_loan_amount, SUM(  t.transaction_fee) AS total_transaction_fee, SUM(t.amount) AS total_transaction_amount
    FROM loans l INNER JOIN transactions t ON l.id = t.loan_id AND l.user_id = t.user_id
GROUP BY 1,2
```

```python
# Show the total number of loans created in a weekday, and order Descending 
```

```python
%%sql
SELECT weekday, SUM(Num_loans_per_day) AS Num_loans_per_weekday, SUM(Total_loan_amount) AS Total_loans_amount

/* Create a table that shows total number of loans created each day , Show its weekday, Order from recent to past*/
FROM(
        SELECT DATE_TRUNC('day', created) AS loan_created , EXTRACT (ISODOW FROM DATE_TRUNC('day', created)) ,
               CASE EXTRACT (ISODOW FROM DATE_TRUNC('day', created)) 
               WHEN 0 THEN 'Su'
               WHEN 1 THEN 'Mo'
               WHEN 2 THEN 'Tu'
               WHEN 3 THEN 'We'
               WHEN 4 THEN 'Th'
               WHEN 5 THEN 'Fr'
               ELSE  'Sa'
               END AS weekday ,COUNT(DISTINCT id) AS Num_loans_per_day, SUM(amount) AS Total_loan_amount
        FROM loans
        GROUP BY 1
        ORDER BY 1 DESC
    ) AS organized_loans
GROUP BY 1
ORDER BY 2 DESC
```
